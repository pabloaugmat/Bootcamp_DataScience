```python
url_dataset02 = "https://raw.githubusercontent.com/pabloaugmat/Bootcamp_DataScience/main/A170724189_28_143_208.csv"

dados = pd.read_csv(url_dataset02, encoding="ISO-8859-1", sep=";", skiprows=3, skipfooter=12, decimal=".", thousands=",", engine="python")

dados
```

```python
dados[["2008/Ago", "2008/Set"]].head()
```

```python
dados.mean().index
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
dados[colunas_usaveis].head()
```

```python
usaveis = dados[colunas_usaveis]
usaveis = usaveis.set_index("Unidade da Federação")
usaveis.head()
```

#Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada



```python
#tratando os dados para melhor apresentação grafica
usaveis = usaveis.drop("Total", axis= 1)
```

```python
ax = usaveis.T.plot(figsize = (12,8))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valores por unidade da federação no periodo de {} a {}".format(dados.columns[1], dados.columns[-2]))
plt.show()
```

#Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```python
ax = usaveis.head().T.plot(figsize = (12,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valores de 5 unidades federais")
plt.show()
```

# Desafio 02: Adicionar uma coluna mostrando a região (Norte, Nodeste, Sul, Suldeste e Centro-Oeste) de cada estado.

```python

regioes = ['','Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']
regioes_indice = []
i = 1
for indice in usaveis.index:
  if indice[0] == str(i):
    regioes_indice.append(regioes[i])
  else:
    i+=1
    regioes_indice.append(regioes[i])

regioes_indice
```

```python
usaveis.insert(0, 'Região', value=regioes_indice)
```

```python
usaveis
```

# Desafio 03: Formatar o gráfico de custos por mês dos 5 estados, deixando ele agradável (Bonitão, segundo o Gui)

```python
#usaveis.drop('Região', axis=1, inplace=True)
usaveis
```

```python
usaveis['Total'] = usaveis.sum(axis=1)
usaveis.head()
```

```python
usaveis.set_index('Região', append=True, inplace=True)
```

```python
usaveis.sort_values('Total', ascending=False, inplace=True)
usaveis.head()
usaveis.drop('Total',axis=1 ,inplace=True)
```

```python
ax = usaveis.head().T.plot(figsize = (12,8),cmap='gist_rainbow')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.xticks(rotation = 45, fontsize = 15)
plt.yticks(fontsize=15)

plt.title("Valores por UF dos 5 estados com maiores gastos",fontsize = 25)
plt.grid()
```

# Desafio 06: Plotar o gráfico dos custos apenas dos estados da região sudeste e verificar se os picos de 2013/Fev teve comportamento similar em todos os demais estados da região

```python
sudeste = []

for indice in usaveis.index.tolist():
  if 'Sudeste' in indice:
    sudeste.append(indice)
```

```python
sudeste
```

```python
ax = usaveis.loc[sudeste].T.plot(figsize = (12,8), cmap = 'gist_rainbow')
plt.legend(labels= (sudeste[0][0],sudeste[1][0],sudeste[2][0],sudeste[3][0]))

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

ax.set_facecolor("#737070")

plt.title("Valores gastos pelos estados do Sudeste \nno período de {} a {}".format(usaveis.columns[0],usaveis.columns[-1]), fontsize=20)

plt.yticks(fontsize=15)
plt.xticks(rotation=45, fontsize = 15)

plt.grid()
plt.show()
```
